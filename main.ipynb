```python
import pandas as pd
# Read the csv file into a pandas dataframe
df = pd.read_csv('spotify_songs.csv')
df.describe().T
```

```python
# Looking at the 1st 5 observations - transposed because we have more columns than rows
df.head().T
```

```python
# Check if track_id is unique for each row
print(f"Number of unique track_ids: {df['track_id'].nunique()}")
print(f"Total number of rows: {len(df)}")
```

```python
# Check for duplicate rows
print(f"Number of duplicate rows: {df.duplicated().sum()}")
```

```python
# Check for missing values in all columns - print results if we get more than 0
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0])
```

```python
# Query to find rows with null values in specified columns
null_rows = df[df['track_name'].isnull() | 
               df['track_artist'].isnull() | 
               df['track_album_name'].isnull()]

# Display the results
# print(f"Number of rows with null values: {len(null_rows)}")
# print("\nRows with null values:")
print(null_rows)
```

```python
# Examine unique values and their counts for categorical columns
for col in ['playlist_genre', 'playlist_subgenre', 'track_artist', 'track_album_name']:
    print(f"\nUnique values in {col}:")
    print(df[col].value_counts().head())
```

```python
# dates appear to be in different formats
format_counts = df['track_album_release_date'].value_counts().head()
print("\nMost common date formats:")
print(format_counts)
```

```python
# parse dates
def parse_date(date_string):
    formats = ["%Y-%m-%d", "%Y-%m", "%Y"]
    for fmt in formats:
        try:
            return pd.to_datetime(date_string, format=fmt)
        except ValueError:
            pass
    return pd.NaT  # Return Not a Time for any unparseable formats

# Apply the custom function
df['track_album_release_date'] = df['track_album_release_date'].apply(parse_date)

# Check the result
print("\nDate column after parsing:")
print(df['track_album_release_date'].describe())

unparsed = df[df['track_album_release_date'].isnull()]
print("\nNumber of unparsed dates:", len(unparsed))
if len(unparsed) > 0:
    print("Sample of unparsed dates:")
    print(unparsed['track_album_release_date'].head())
```

```python
# Analyze the Parsed Dates
print("\nDistribution of album release years:")
year_counts = df['track_album_release_date'].dt.year.value_counts().sort_index()
print(year_counts)

# Visualize the distribution
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
year_counts.plot(kind='bar')
plt.title('Distribution of Album Release Years')
plt.xlabel('Year')
plt.ylabel('Number of Albums')
plt.xticks(rotation=45)
plt.show()
```
